6. (7б) Визначте функцію, яка приймає iterable (ітерабельний обʼєкт, напр. список), що складається з 7ми рядкових значень - назв погодних умов (напр. "sunny") за кожен день у тижні. Повертає функція словник, де ключами є рядки з погодними умовами, а значеннями - кількість разів, скільки дана погода була протягом тижня (напр. {'sunny': 1, ...}). Викличте цю функцію та виведіть результат її роботи на екран.

```python
def count_weather(weather_list):
  weather_counts = {}
  for condition in weather_list:
    if condition in weather_counts:
      weather_counts[condition] += 1
    else:
      weather_counts[condition] = 1
  return weather_counts

weather_conditions = ["sunny", "cloudy", "windy", "rainy", "windy", "sunny", "foggy"]
result = count_weather(weather_conditions)
print(result)
```

7. (15б) Напишіть програму, яка б шукала необхідне число у впорядкованому списку цілочислених значень, де
число має вводитись з консолі.
власне алгоритм пошуку винесений в окрему функцію, яка повертає булеве значення (знайдено число - True, не знайдено - False).
в залежності від результату функції, виведіть на екран повідомлення про успішний або ж неуспішний пошук.
Зауваження: використовувати оператор in заборонено.

```python
def is_num_in_list(target, list_of_nums):
  for num in list_of_nums:
    if num == target:
      return True
  return False


list_of_nums = [0, 1, 2, 3, 4, 5, 6, 7, 8]
find_num = int(input("Введіть шукане число: "))

if(is_num_in_list(find_num, list_of_nums)):
  print(f"Число {find_num} є в списку")
else:
  print(f"Число {find_num} немає в списку")
```

# III. Decorators.

11. (11б) Визначте декоратор, який би виводив у консоль повідомлення про початок роботи і назву функції (напр. my_function._ _ name _ _), а після відпрацювання функції - повідомлення про закінчення роботи і назву функції, що задекорована.

```python
def my_decorator11(func):
  def wrapper(*args, **kwargs):
    print(f"Початок роботи {func.__name__}")
    result = func(*args, **kwargs)
    print(f"Кінець роботи {func.__name__}")
    return result
  return wrapper

@my_decorator11
def greet():
  print("Hello")

greet()
```

12. (4б) Задекоруйте двома різними способами функції з завдання 4 та 5 декоратором із завдання 11. (**УВАГА**: не декоруйте функції двічі!).

```python
def my_decorator12(func):
  def wrapper(*args, **kwargs):
    print(f"Початок роботи {func.__name__}")
    result = func(*args, **kwargs)
    print(f"Кінець роботи {func.__name__}")
    return result
  return wrapper


@my_decorator12
def my_divicion4(arg1, arg2):
  if (arg2 == 0):
    return None
  return arg2 / arg1


def my_subtraction(arg1, arg2=1):
  return arg1 - arg2
my_subtraction = my_decorator12(my_subtraction)
```

Наприклад, потрібно знайти перше число у послідовності, тоді функція прийматиме 0 у якості аргумента і повертатиме 1, якщо ж знаходити третє число (індекс 2) послідовності, то функція повертатиме 2, якщо ж знаходити десяте число (індекс 9), то функція повертатиме 55.

```python
def fibonachi_of(n):
  if n==0 or n==1:
    return 1
  return fibonachi_of(n-1) + fibonachi_of(n-2)

print(fibonachi_of(2))
```
