15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

**Схожості:**
- Обидва типи зберігають елементи у чітко визначеному порядку.
- Доступ до елементів здійснюється за індексами, підтримуються зрізи.
- Можуть містити елементи різних типів одночасно.
- Дозволяють зберігати однакові значення

**Відмінності:**

*Список(List):*
- Створення за допомогою []
- Будь-які зміни модифікують вже створений об'єкт (mutable)
- Повільний

*Кортеж(Tuple):*
- Створення за допомогою ()
- Будь-які зміни створюють новий об'єкт (immutable)
- Швидкий


**Випадки використання:**

*Список(List):*
- Коли планується змінювати дані (додавати, видаляти, сортувати)
- Для колекцій однотипних об'єктів, кількість яких заздалегідь невідома

*Кортеж(Tuple):*
- Для сталих даних, які не повинні змінюватися протягом роботи програми
- Коли потрібна висока продуктивність
- Коли потрібно використовувати колекцію як ключ у словнику.

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
my_set.add(3.14)
my_set.add((1, 2))
my_set.update(["Hello", 100, True])
print(my_set)
```

Так. Mutable: You can add or remove elements from an existing set.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_new_set = set(my_list)
print(my_new_set)
```

Тому що set -- це множина з унікальними елементами. Тобто об'єкт створюється без дублікатів

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
my_set1 = {1, 2, 3, 4, 5}
my_set2 = {3, 4, 5, 6, 7}

union_sets = my_set1.union(my_set2)     #union = всі елементи з обох множин
intersection_sets = my_set1.intersection(my_set2)     #intersection = тільки спільні елементи
difference_sets = my_set1.difference(my_set2)    #difference = елементи, що є в A, але немає в B
sym_diff_sets = my_set1.symmetric_difference(my_set2)   #symmetric_difference = елементи, що не є спільними

print(union_sets)
print(intersection_sets)
print(difference_sets)
print(sym_diff_sets)
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
my_dict = dict()
my_dict.update({0 : False, 3.14 : "pi", True : 1, (1,2,3) : "tuple"})
print(my_dict)
```

Список (list) не може бути ключем, оскільки списки є змінними (mutable) об'єктами. А ключі в словниках повинні бути тільки незмінними (immutable), щоб забезпечити стабільність хеш-таблиці!
