5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure3D:
  def get_volume(self):
    return 0

  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math
class Cylinder(Figure3D):
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def get_volume(self):
    return math.pi * (self.radius ** 2) * self.height


class Cube(Figure3D):
  def __init__(self, edge):
    self.edge = edge

  def get_volume(self):
    return self.edge ** 3


figure1 = Cylinder(3, 4)
figure2 = Cube(3)
total_volume = figure1 + figure2

print(f"Об'єм циліндра: {figure1.get_volume():.2f}")
print(f"Об'єм куба: {figure2.get_volume()}")
print(f"Сумарний об'єм: {total_volume:.2f}")
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, balance, id):
    self.__balance = balance   #private
    self.__id = id             #private

  @property                    #геттер (просто перегляд балансу)
  def balance(self):
    return self.__balance

  @property
  def id(self):
    return self.__id

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Рахунок поповнено на {amount}. Баланс: {self.__balance}")
    else:
        print("Неможливо виконати операцію")

  def withdraw(self, amount):
    if 0 < amount <= self.__balance:
        self.__balance -= amount
        print(f"Знято {amount}. Залишок: {self.__balance}")
    else:
        print("Некоректна сума")

wallet = EWallet(15, 1)
print(f"Гаманець {wallet.id}, Баланс: {wallet.balance}")
wallet.deposit(20)
wallet.withdraw(10)
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name):
    self.name = name

class ShoppingCart:
  def __init__(self, items):
    self.items = items

  def show_items(self):
    if not self.items:
        print("Кошик порожній")
        return
    res = ", ".join([item.name for item in self.items])
    print(f"В кошику: {res}")

  def add_item(self, new_item):
    self.items.append(new_item)
    print(f"{new_item.name} було додано")

  def delete_item(self, item_to_delete):
    for item in self.items:
      if item.name == item_to_delete:
        self.items.remove(item)
        print(f"{item_to_delete} було видалено")
        return
    print(f"{item_to_delete} не знайдено")


our_list = list((Item("Pen"), Item("Ball")))
cart = ShoppingCart(our_list)
cart.show_items()
cart.add_item(Item("Banana"))
cart.show_items()
cart.delete_item("Ball")
cart.show_items()
cart.delete_item("Milk")
```

10. (13б). Створіть клас Vehicle з атрибутами brand (manufacturer) та production_age. Створіть ще один клас Engine з такими атрибутами, як type (ex. petrol) та size (volume, litres). Створіть клас Car, який успадковує обидва класи Vehicle та Engine. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Car та отримавши mro для цього класу.

```python
class Vehicle:
  def __init__(self, brand, production_age):
    self.brand = brand
    self.production_age = production_age

class Engine:
  def __init__(self, engine_type, engine_size):
    self.engine_type = engine_type
    self.engine_size = engine_size

class Car(Vehicle, Engine):
  def __init__(self, brand, production_age, engine_type, engine_size):
    Vehicle.__init__(self, brand, production_age)
    Engine.__init__(self, engine_type, engine_size)

car1 = Car("Nissan", 2019, "petrol", 1.3)
print(car1.brand)
print(car1.production_age)
print(car1.engine_type)
print(car1.engine_size)
print(Car.mro())
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
  def __init__(self):
    self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    self.index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.index < len(self.days):
      day = self.days[self.index]
      self.index += 1
      return day
    else:
      raise StopIteration

week = WeekDays()

for day in week:
  print(day)
```
